```python
#contour par sobel
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from google.colab import drive

# Monter Google Drive
drive.mount('/content/drive')

# Chemin vers le dossier contenant les images
input_folder = '/content/drive/MyDrive/data/fichier  filtrer yes'  # Modifier avec le chemin de votre dossier
# Chemin vers le dossier pour sauvegarder les résultats
output_folder = '/content/drive/MyDrive/data/IRM_Sobel_results'
os.makedirs(output_folder, exist_ok=True)  # Créer le dossier de sortie s'il n'existe pas

# Parcourir toutes les images dans le dossier
for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png') or filename.endswith('.jpeg'):
        file_path = os.path.join(input_folder, filename)

        # Charger l'image en niveaux de gris
        image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
        if image is None:
            print(f"Erreur de chargement pour l'image : {filename}")
            continue

        # Appliquer un filtre pour réduire le bruit
        filtered_image = cv2.GaussianBlur(image, (5, 5), 0)  # Utilisez le filtre gaussien ici

        # Calcul des gradients Sobel dans les directions x et y
        sobel_x = cv2.Sobel(filtered_image, cv2.CV_64F, 1, 0, ksize=3)
        sobel_y = cv2.Sobel(filtered_image, cv2.CV_64F, 0, 1, ksize=3)

        # Magnitude des gradients
        sobel_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
        sobel_magnitude = np.uint8(np.clip(sobel_magnitude, 0, 255))

        # Normalisation pour améliorer la visualisation
        sobel_normalized = cv2.normalize(sobel_magnitude, None, 0, 255, cv2.NORM_MINMAX)

        # Sauvegarder l'image résultat
        output_path = os.path.join(output_folder, f"Sobel_{filename}")
        cv2.imwrite(output_path, sobel_normalized)

        # Optionnel : Affichage des résultats
        plt.figure(figsize=(12, 6))

        plt.subplot(1, 2, 1)
        plt.title('Image originale')
        plt.imshow(image, cmap='gray')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.title('Contours détectés (Sobel)')
        plt.imshow(sobel_normalized, cmap='gray')
        plt.axis('off')

        plt.tight_layout()
        plt.show()

print(f"Les contours détectés ont été sauvegardés dans : {output_folder}")
```

```python
#contour avec canny
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from google.colab import drive

# Monter Google Drive
drive.mount('/content/drive')

# Chemin vers le dossier contenant les images
input_folder = '/content/drive/MyDrive/data/fichier  filtrer yes'  # Modifier avec le chemin de votre dossier
# Chemin vers le dossier pour sauvegarder les résultats
output_folder = '/content/drive/MyDrive/data/IRM_Canny_results'
os.makedirs(output_folder, exist_ok=True)  # Créer le dossier de sortie s'il n'existe pas

# Paramètres pour l'algorithme de Canny
low_threshold = 50  # Seuil inférieur
high_threshold = 150  # Seuil supérieur

# Parcourir toutes les images dans le dossier
for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png') or filename.endswith('.jpeg'):
        file_path = os.path.join(input_folder, filename)

        # Charger l'image en niveaux de gris
        image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
        if image is None:
            print(f"Erreur de chargement pour l'image : {filename}")
            continue

        # Appliquer un filtre pour réduire le bruit
        filtered_image = cv2.GaussianBlur(image, (5, 5), 0)

        # Appliquer l'algorithme de Canny
        edges = cv2.Canny(filtered_image, low_threshold, high_threshold)

        # Sauvegarder l'image résultat
        output_path = os.path.join(output_folder, f"Canny_{filename}")
        cv2.imwrite(output_path, edges)

        # Optionnel : Affichage des résultats
        plt.figure(figsize=(12, 6))

        plt.subplot(1, 2, 1)
        plt.title('Image originale')
        plt.imshow(image, cmap='gray')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.title('Contours détectés (Canny)')
        plt.imshow(edges, cmap='gray')
        plt.axis('off')

        plt.tight_layout()
        plt.show()

print(f"Les contours détectés ont été sauvegardés dans : {output_folder}")
```

```python
#contour et surface
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from google.colab import drive

# Monter Google Drive
drive.mount('/content/drive')

# Chemin vers le dossier contenant les images
input_folder = '/content/drive/MyDrive/data/fichier  filtrer yes'
# Chemin vers le dossier pour sauvegarder les images filtrées
output_folder = '/content/drive/MyDrive/data/fichier_contour yes'  # Dossier de sortie

# Créer le dossier de sortie s'il n'existe pas
os.makedirs(output_folder, exist_ok=True)

# Étape 2 : Charger et traiter toutes les images
for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Vérifiez les extensions
        # Charger l'image
        file_path = os.path.join(input_folder, filename)
        image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

        if image is not None:
            # Appliquer un seuillage simple pour la segmentation de la tumeur
            _, thresholded_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

            # Détection des contours
            contours, _ = cv2.findContours(thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

            # Initialiser une variable pour stocker la surface totale de la tumeur
            total_area = 0

            # Parcourir les contours détectés
            for contour in contours:
                # Calculer l'aire de chaque contour
                area = cv2.contourArea(contour)
                if area > 100:  # Seuil pour ignorer les petits contours
                    total_area += area

                    # Dessiner les contours sur l'image originale
                    cv2.drawContours(image, [contour], -1, (0, 255, 0), 2)  # Dessine en vert

            # Afficher l'image avec les contours détectés et l'aire
            plt.figure(figsize=(10, 5))
            plt.subplot(1, 2, 1)
            plt.title(f'Contours de la Tumeur: {filename}')
            plt.imshow(image, cmap='gray')
            plt.axis('off')

            # Affichage de l'aire totale de la tumeur
            print(f"Image: {filename}, Surface totale de la tumeur: {total_area} pixels")

            # Sauvegarder l'image avec les contours dessinés
            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, image)

            # Affichage des résultats
            plt.subplot(1, 2, 2)
            plt.title(f"Surface Tumeur: {total_area} pixels")
            plt.imshow(thresholded_image, cmap='gray')
            plt.axis('off')

            plt.show()

print("Les images avec les contours de la tumeur ont été sauvegardées dans :", output_folder)
```

```python
#emplacement:
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from google.colab import drive

# Monter Google Drive
drive.mount('/content/drive')

# Chemin vers le dossier contenant les images
input_folder = '/content/drive/MyDrive/data/fichier_contour yes'
# Chemin vers le dossier pour sauvegarder les images filtrées
#output_folder = '/content/drive/MyDrive/data/fichier_filtré_no'  # Dossier de sortie

# Créer le dossier de sortie s'il n'existe pas
#os.makedirs(output_folder, exist_ok=True)

# Définir les zones du cerveau (gauche, droite, supérieur, inférieur)
def localize_tumor(x, y, width, height):
    """Classifie l'emplacement de la tumeur (gauche, droite, supérieur, inférieur)."""
    if x < width // 2:
        horizontal = 'Gauche'
    else:
        horizontal = 'Droite'

    if y < height // 2:
        vertical = 'Supérieur'
    else:
        vertical = 'Inférieur'

    return f"{vertical}-{horizontal}"

# Étape 2 : Charger et traiter toutes les images
for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Vérifiez les extensions
        # Charger l'image
        file_path = os.path.join(input_folder, filename)
        image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

        if image is not None:
            # Appliquer un seuillage simple pour la segmentation de la tumeur
            _, thresholded_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

            # Détection des contours
            contours, _ = cv2.findContours(thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

            # Initialiser une variable pour stocker l'emplacement de la tumeur
            tumor_location = ""

            # Parcourir les contours détectés
            for contour in contours:
                # Calculer l'aire de chaque contour
                area = cv2.contourArea(contour)
                if area > 100:  # Seuil pour ignorer les petits contours
                    # Calculer les coordonnées du centre du contour
                    M = cv2.moments(contour)
                    if M["m00"] != 0:
                        cX = int(M["m10"] / M["m00"])
                        cY = int(M["m01"] / M["m00"])

                        # Localiser la tumeur dans l'image
                        tumor_location = localize_tumor(cX, cY, image.shape[1], image.shape[0])

                        # Dessiner les contours et le centre de la tumeur
                        cv2.drawContours(image, [contour], -1, (0, 255, 0), 2)  # Dessine en vert
                        cv2.circle(image, (cX, cY), 7, (0, 0, 255), -1)  # Centre de la tumeur en rouge

            # Afficher l'image avec les contours et l'emplacement de la tumeur
            plt.figure(figsize=(10, 5))
            plt.subplot(1, 2, 1)
            plt.title(f'Contours de la Tumeur: {filename}')
            plt.imshow(image, cmap='gray')
            plt.axis('off')

            # Affichage de l'emplacement de la tumeur
            print(f"Image: {filename}, Emplacement de la tumeur: {tumor_location}")

            # Sauvegarder l'image avec les contours dessinés
            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, image)

            # Affichage de l'image avec les contours et le centre
            plt.subplot(1, 2, 2)
            plt.title(f"Emplacement: {tumor_location}")
            plt.imshow(thresholded_image, cmap='gray')
            plt.axis('off')

            plt.show()

print("Les images avec les contours et l'emplacement de la tumeur ont été sauvegardées dans :", output_folder)
```

```python
# segmentation k means
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from google.colab import drive

# Monter Google Drive
drive.mount('/content/drive')

# Chemin vers le dossier contenant les images
input_folder = '/content/drive/MyDrive/data/fichier_contour yes'
# Chemin vers le dossier pour sauvegarder les images segmentées
output_folder = '/content/drive/MyDrive/data/fichier_segmenté_yes'  # Dossier de sortie

# Créer le dossier de sortie s'il n'existe pas
os.makedirs(output_folder, exist_ok=True)

# Fonction pour appliquer K-means sur une image
def kmeans_segmentation(image, k=2):
    """Applique la segmentation par K-means sur l'image"""
    # Reshaper l'image pour appliquer K-means sur chaque pixel
    pixel_values = image.reshape((-1, 1))
    pixel_values = np.float32(pixel_values)

    # Critères d'arrêt et nombre de clusters
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
    _, labels, centers = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

    # Convertir les centres en uint8
    centers = np.uint8(centers)
    segmented_image = centers[labels.flatten()]
    segmented_image = segmented_image.reshape(image.shape)

    return segmented_image, labels.reshape(image.shape)

# Étape 2 : Charger et traiter toutes les images
for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Vérifiez les extensions
        # Charger l'image
        file_path = os.path.join(input_folder, filename)
        image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

        if image is not None:
            # Appliquer la segmentation par K-means
            k = 2  # Nombre de clusters à détecter
            segmented_image, labels = kmeans_segmentation(image, k)

            # Trouver le cluster correspondant à la tumeur (selon l'intensité des pixels)
            # Supposons que la tumeur correspond au cluster avec la plus grande intensité
            unique, counts = np.unique(labels, return_counts=True)
            max_cluster = unique[np.argmax(counts)]  # Cluster majoritaire
            tumor_segment = (labels == max_cluster).astype(np.uint8) * 255

            # Afficher les résultats
            plt.figure(figsize=(15, 10))

            plt.subplot(2, 2, 1)
            plt.title(f'Image Originale: {filename}')
            plt.imshow(image, cmap='gray')
            plt.axis('off')

            plt.subplot(2, 2, 2)
            plt.title('Image Segmentée par K-means')
            plt.imshow(segmented_image, cmap='gray')
            plt.axis('off')

            plt.subplot(2, 2, 3)
            plt.title('Segmentation de la Tumeur')
            plt.imshow(tumor_segment, cmap='gray')
            plt.axis('off')

            plt.show()

            # Sauvegarder l'image segmentée
            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, tumor_segment)

print("Les images segmentées par K-means ont été sauvegardées dans :", output_folder)
```

```python
#reconstruction 3d
# Imports
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
import plotly.graph_objects as go

# Chemins des images
image_paths = [
   "/content/drive/MyDrive/data/Copie de Y8.jpg",
    "/content/drive/MyDrive/data/Copie de Y8.jpg",
    "/content/drive/MyDrive/data/Copie de Y8.jpg"
]

# Charger les images
images = []
for path in image_paths:
    image = Image.open(path).convert('RGB')  # Charger en RGB
    images.append(np.array(image))

# Prétraitement : conversion en niveaux de gris
gray_images = [cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) for img in images]

# Application du seuil d'Otsu pour la segmentation
binary_images = []
for gray_image in gray_images:
    thresh_value = threshold_otsu(gray_image)
    binary_images.append(gray_image > thresh_value)

# Normalisation et création de cartes de profondeur pour chaque image
z_scale = 0.3  # Échelle de profondeur
depth_maps = [binary_image.astype(np.float32) * z_scale for binary_image in binary_images]

# Création de la grille (X, Y) pour l'affichage
x = np.linspace(0, 1, gray_images[0].shape[1])  # Coordonnées X
y = np.linspace(0, 1, gray_images[0].shape[0])  # Coordonnées Y
x, y = np.meshgrid(x, y)

# Assemblage des couches en une seule surface pour la visualisation 3D
combined_depth = np.stack(depth_maps, axis=-1).mean(axis=-1)  # Moyenne des cartes de profondeur

# Visualisation en 3D
fig = go.Figure(data=[go.Surface(
    z=combined_depth,
    x=x,
    y=y,
    surfacecolor=gray_images[0],  # Utiliser une des images comme texture
    colorscale='Gray',
    cmin=0,
    cmax=255
)])

# Configurer les axes et les options de la scène
fig.update_layout(scene=dict(
    xaxis_title='X',
    yaxis_title='Y',
    zaxis_title='Profondeur',
    zaxis=dict(range=[0, z_scale])  # Plage de profondeur
))

# Afficher la figure 3D
fig.show()
```
